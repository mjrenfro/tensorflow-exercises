```python

%matplotlib inline

import numpy as np
import skimage.io as io

cat_img = io.imread('resources/cats.jpg')
io.imshow(cat_img)
```

```python
#Convert the picture to a string rep
cat_string = cat_img.tostring()

#Convert the string back to the image

#First get the 1d representation
reconstructed_cat_1d=np.fromstring(cat_string, dtype=np.uint8)

#Reshape the 1d representation according to the previously
#defined image
reconstructed_cat_img=reconstructed_cat_1d.reshape(cat_img.shape)

#Compare with the original
np.allclose(cat_img, reconstructed_cat_img)
```
